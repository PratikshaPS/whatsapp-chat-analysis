```python
import re
import pandas as pd
```

```python
file = open('WhatsApp Chat with iNeuron achiever.txt','r',encoding='utf-8')
```

```python
data = file.read()
```

```python
print(data)
```

```python
len(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s[apAP][mM]\s-\s'
```

```python
date = re.findall(pattern, data)
date = [dte.replace('\u202f'," ") for dte in date]
```

```python
len(date)
```

```python
messages = re.split(pattern, data)[1:]
```

```python
len(messages)
```

```python
messages
```

```python
df = pd.DataFrame({'user_message':messages, 'message_date':date})

df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%Y, %H:%M %p - ')

df.rename(columns={'message_date':'date'}, inplace=True)

df.head()
```

```python
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:  # user name
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group_notification')
        messages.append(entry[0])
```

```python
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)
```

```python
df
```

```python
df['only_date'] = df['date'].dt.date
df['year'] = df['date'].dt.year
df['month_num'] = df['date'].dt.month
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['day_name'] = df['date'].dt.day_name()
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
```

```python
temp = df[df['user'] != 'group_notification']
temp = temp[temp['message'] != '<Media omitted>\n']
```

```python
temp
```

```python
f = open('stop_hinglish.txt', 'r')
stop_words = f.read()
print(stop_words)
```

```python
words = []

for message in temp['message']:
    for word in message.lower().split():
        if word not in stop_words:
            words.append(word)
```

```python
words
```

```python
from collections import Counter
```

```python
most_common_df = pd.DataFrame(Counter(words).most_common(20))
```

```python
most_common_df
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()

ax.barh(most_common_df[0], most_common_df[1])
ax.invert_yaxis()
plt.xticks(rotation='vertical')
```
